```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import zscore
from sklearn.cluster import AgglomerativeClustering
import os
import math
```

```python
dfs_dni = []
dfs_wind = []
for root, dirs, files in os.walk('.'):
    for name in files:
        if name.endswith('.csv') :

            if name == 'Delhi_2020.csv' :
                continue
            file = os.path.join(root , name)
            df = pd.read_csv(file , header =10)
            dfs_dni.append(df['ALLSKY_SFC_LW_DWN'])
            dfs_wind.append(df['WS50M'])




        
        
            
            
```

```python
dni = pd.concat(dfs_dni , axis = 1  )
wind = pd.concat(dfs_wind , axis = 1 )
dni.columns = ['CD' , 'ED' , 'NEWD' , 'ND' , 'NED' , 'NWD' , 'SD' , 'SED' , 'SWD' , 'WD']
wind.columns = ['CD' , 'ED' , 'NEWD' , 'ND' , 'NED' , 'NWD' , 'SD' , 'SED' , 'SWD' , 'WD']
```
